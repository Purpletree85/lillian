```python
import geopandas as gpd
import folium 
```

```python
lillian = gpd.read_file("C:/Users/lgadd/Downloads/lillian.shp")
lillian_json = lillian.to_crs(epsg='4326').to_json()
```

```python
# NYC bounds
nyc_coords = [40.7128, -74.0060]
nyc_bounds = [[40.4774, -74.2591], [40.9176, -73.7004]]
m = folium.Map(
    location=nyc_coords,
    zoom_start=12,
    min_zoom=11,
    max_zoom=18,
    maxBounds=nyc_bounds,  # Fixed: pass the bounds array
    control_scale=True,
    tiles="OpenStreetMap"
)
```

```python
folium.GeoJson(
    lillian_json,
    style_function=lambda feature: {
        'fillColor': '#ff69b4',
        'color': 'black',
        'weight': 1,
        'fillOpacity': 0.5,
        'radius': 5,  # Important for points
        'fill': True  # Important for points
    },
    marker_options=folium.vector_layers.CircleMarker(radius=5),  # Explicit marker
    tooltip=folium.features.GeoJsonTooltip(
        fields=['COMPANY','LOCATION'],
        aliases=['Company:', 'Location:'],
        localize=True
    )
).add_to(m)
m
```
